# Dependencies

```python
import mediapipe as mp
import cv2
import numpy as np
import uuid
import os
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
```

# Drawing the Hands

#### the image below shows how using mediapipe we can identify the joints in our hands

<img src=https://i.imgur.com/qpRACer.png />

```python
# capturing the live video feed using opencv

camcap = cv2.VideoCapture(0)

with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands:
    while camcap.isOpened(): 
        ret, frame = camcap.read()
        
        # color conversion:
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # fliping image horizontally
        image = cv2.flip(image, 1)
        
        #setting flag to false
        image.flags.writeable=False
        
        # making dectections
        results = hands.process(image)
        
        # setting flag to true
        image.flags.writeable = True
        
        # reverting the color converstion
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # printing detections.
        print(results)
        
        # rendering results
        if results.multi_hand_landmarks:
            for num, hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS, 
                                         mp_drawing.DrawingSpec(color=(121, 22, 76), thickness=2, circle_radius=4),
                                         mp_drawing.DrawingSpec(color=(255, 53, 10), thickness=2, circle_radius=2),)
        

        cv2.imshow('Hand Tracking', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

camcap.release()
cv2.destroyAllWindows()
```

```python

# we will wrtie a function to map through the multi_handeness to the actual hand that is being detected

def getLabel(index, hand, results):
    output = None
    for idx, classification in enumerate(results.multi_handedness):
        if classification.classification[0].index == index:
            
            # processing the results
            label = classification.classification[0].label
            score = classification.classification[0].score
            text = '{} {}'.format(label, round(score, 2))
            
            # extracting coordinates
            coor = tuple(np.multiply(
                np.array((hand.landmark[mp_hands.HandLandmark.WRIST].x, hand.landmark[mp_hands.HandLandmark.WRIST].y)),
            [640,480]).astype(int))
            
            output = text, coor
            
    return output
```

# Calculate Joint Angles

```python
from matplotlib import pyplot as plt
```

```python
jointList = [[8,7,6], [12, 11, 10], [4, 3, 2], [16, 15, 14], [20, 19, 18]]
```

```python
# function to calculate and draw angles on joints

def drawJointAngles(images, results, jointList):
    
    # looping through the hands arrays
    for hand in results.multi_hand_landmarks:
        # looping through jointList
        for joint in jointList:
            a = np.array([hand.landmark[joint[0]].x, hand.landmark[joint[0]].y]) # first coordinate for first instance in joint list
            b = np.array([hand.landmark[joint[1]].x, hand.landmark[joint[1]].y]) # second coordinate
            c = np.array([hand.landmark[joint[2]].x, hand.landmark[joint[2]].y]) # thrid coordinate
            
            # calculate angle
            radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]- b[0])
            
            # converting radians into degrees
            angle = np.abs(radians*180.0/np.pi)
            
            if angle > 180.0:
                angle = 360-angle
                
            cv2.putText(image, str(round(angle, 2)), tuple(np.multiply(b, [640, 480]).astype(int)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
    return image
    
```
